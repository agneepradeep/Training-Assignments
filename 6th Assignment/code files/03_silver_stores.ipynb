```python
from pyspark.sql.functions import col
```

# Paths

```python
bronze_stores_path = "dbfs:/Volumes/workspace/retail/raw/bronze/stores/stores.csv"
```

# Read Bronze Stores (CSV)

```python
stores_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(bronze_stores_path)
)
```

# Basic cleansing

```python
silver_stores_df = (
    stores_df
    .dropDuplicates(["store_id"])
    .filter(col("store_id").isNotNull())
)
```

# Write Silver Stores (overwrite snapshot)

```python
(
    silver_stores_df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("workspace.retail.silver_stores")
)
```
